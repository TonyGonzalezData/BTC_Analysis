First, we will focus the analysis solely on asset, date and kline data (open, high, low, and close), so we need to discard the remaining fields.

```python
historical_data= historical_data[['asset','date','open','high','low','close']]
historical_data
```

### Variables and types

**We review the variable types.**

```python
historical_data.info()
```


**Conclusion**: we convert the following variables:

Date: date

Float: open, high, low, close

```python
for variable in ['open','high','low','close']:
    historical_data[variable] = historical_data[variable].astype('float')
```

```python
historical_data.date = pd.to_datetime(historical_data.date)
```

Check

```python
historical_data.info()
```

#### Analysis of nulls

From the Non-null column of the info(), we see that we have no nulls.

#### Duplicated analysis

We check if there is some duplicated register.

```python
historical_data.duplicated().sum()
```

#### Analysis of datetime variables

We need to ensure there are no gaps in the dates for each asset.

```python
historical_data['gap'] = historical_data[['asset', 'date']].sort_values(by=['asset', 'date']).groupby('asset').diff() > pd.to_timedelta('1 day')
historical_data[historical_data.gap]
```

There are no gaps. So, we can drop de column.

```python
historical_data.drop(columns='gap',inplace=True)
historical_data
```

#### Analysis of categorical variables

We are going to analyze the values and frequencies of the categorical variables.

```python
historical_data.select_dtypes('O')
```

```python
historical_data.asset.sort_values().value_counts(sort=False)
```

We need to check if the lower counts are related to "young assets." To do this, we’ll examine the time range between the first and last quotation dates for each asset to see if there’s any correlation.

```python
#begining=datetime.today()-timedelta(days=number_of_days-1)
#end=datetime.today()

temp=historical_data.groupby('asset')['date'].agg(['min','max'])

temp['days']=(temp['max']-temp['min']).dt.days+1
temp
```

The last day’s date is the same for all assets, so the time span from the start of quotations to the end matches perfectly with the total record count for each asset.

#### Analysis of numerical variables

```python
historical_data.select_dtypes(include='number').describe().T
```

Conclusions:

* The maximum values need to be checked, as they are likely related to BTC. We should apply the describe function, grouped by asset, to confirm this.

```python
historical_data.groupby('asset')[['open','high','low','close']].max()
```

As we suspected, the maximum values are related to BTC, confirming that this is normal behavior.
